```python
''' Observations from EDA
    - females more likely to churn (+16% compared to males)
    - older more likely to churn as well
    - monthly users have 100% historical churn rate (assuming this company discontinued monthly or does not have recurring payments set up for monthly)
    - 56% of total users churn
    - Older folks more likely to churn
    - More support calls appears to lead to greater churn likelihood, same for payment delay and last interaction
    - Higher spend appears to lead to lower churn likelihood
    - Distributions look to be almost like a piecewise uniform distribution for many cols'''


''' Next steps
    - remove Contract Length from the predictive model.  It seems that there is a major data flaw with monthly users having a 100% churn rate
    on 80,000+ observations.  This is highly unlikely, near impossible.  Data entry error is much more likely.  Since quarterly and yearly users are
    within 0.05% of each other when comparing churn, we will remove the entire variable from the model.  We choose to do this because quarterly vs 
    yearly doesn't seem to make a big difference when predicting churn just based off of Contract Length variable, and because of the massive
    data problem with monthly being 100% churn.
    - covariance matrix / correlation matrix'''
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

from sklearn.model_selection import cross_validate
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
# setup the data for modeling

X = df.drop(columns=['Churn'], axis=1)
y = df['Churn']
```

```python
# encode cat cols
cat_cols.remove('Contract Length')
num_cols.remove('Churn')
encoder = OneHotEncoder(sparse_output=False)
encoder.set_output(transform="pandas")

df_ohe = encoder.fit_transform(X[cat_cols])
X_enc = pd.concat(
    [X[num_cols], df_ohe],
    axis=1
)

X_enc.head
```

```python
# scale numerical values
scaler = StandardScaler()

X_scaled = X_enc.copy()
X_scaled[num_cols] = scaler.fit_transform(X[num_cols])
X_scaled.head
```

```python
# split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=.2, random_state=42)
```

```python
# dictionary of models to fit data to (log reg, rf, xgboost classifier, adaboost classifier, knn)
models = {
    'LogisticRegression': [LogisticRegression(), 
                          {'C': [1, 10], 'max_iter': [500, 1000]}
                          ],
    'RandomForest': [RandomForestClassifier(),
                    {'n_estimators': [10, 50, 100], 'max_depth': [1, 3, 5]}
                    ],
    'GradientBoostingClassifier': [GradientBoostingClassifier(), 
                                  {'n_estimators': [10, 50], 'max_depth': [3, 5]}
                                  ],
    'AdaBoostClassifier': [AdaBoostClassifier(), 
                          {'n_estimators': [25, 75]}
                          ],
    'KNeighborsClassifier': [KNeighborsClassifier(),
                            {'n_neighbors': [3, 5, 10]}]
}
```

```python
metrics_cols = ['model_name','test_accuracy','test_precision','test_recall','test_f1']

model_name=[]
test_accuracy=[]
test_precision=[]
test_recall=[]
test_f1=[]
best_params = []

# metrics for cv
scoring = ['accuracy', 'precision', 'recall', 'f1']

for name, [model, params] in models.items():
    grid = GridSearchCV(model, params, scoring='accuracy', cv=5)

    grid.fit(X_scaled, y)

    best = grid.best_estimator_
    results = cross_validate(best, X_scaled, y, scoring=scoring, cv=5)

    model_name.append(name)
    test_accuracy.append(round(results['test_accuracy'].mean(), 3) * 100)
    test_precision.append(round(results['test_precision'].mean(), 3) * 100)
    test_recall.append(round(results['test_recall'].mean(), 3) * 100)
    test_f1.append(round(results['test_f1'].mean(), 3) * 100)
    best_params.append(grid.best_params_)

    print(f'done with model {name}')
```

```python

metrics_dict = {
    'model': model_name,
    'test_accuracy': test_accuracy,
    'test_precision': test_precision,
    'test_recall': test_recall,
    'test_f1': test_f1,
    'best_params': best_params
}

model_metrics = pd.DataFrame(metrics_dict).sort_values('test_accuracy', ascending=False)

model_metrics
```

```python
final_model = GradientBoostingClassifier(n_estimators=50, max_depth=5)
final_model.fit(X_train, y_train)
y_pred = final_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Test accuracy_score: {accuracy:.2f}")
print(f"Test f1 Score: {f1:.2f}")
```

```python
# confusion matrix
cm = confusion_matrix(y_test, y_pred, labels=final_model.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=final_model.classes_)
disp.plot()
plt.title('Confusion Matrix of Random Forest Classifier')
plt.show()
```
